# 案例分析:  北京房价数据分析

本案例中我们将针对2012年欧洲杯数据进行分析，训练Pandas库的使用

### 导入数据

#### 导入所需的 `pandas` 库

```python
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
pd.set_option('max_columns',100)
```

#### 使用 `pd.read_csv` 导入数据

```python
data = pd.read_csv('lianjia.csv',encoding='gbk')
data
```

房子的建造时间都有哪些？

建造时间最早的房子卖出去多少钱？

```python
data[data['Construction time']=='1997'][['Construction time','Total price']].head()
```

哪个区卖出的房最多？

```python
district_mean=data["District"].value_counts().sort_values(ascending=False)
district_mean
```

距离地铁近的房子卖出去多少套？

```python
data['Subway'].value_counts()
```

海淀区的房价平均多少钱一平米？

```python
data[data['District']=='海淀']['Price'].mean()
```

各种装修情况的房子的平均价格是多少？

```python
data['renovation condition'].value_counts()
```

西城区的房子中有多少两居室

```python
data[(data['District'] == '西城') & (data['Bed Room']==2)].shape[0]
```

100平米以上的房子中，几居室最多

```python
data['Bed Room']=data['Bed Room'].apply(lambda x:str(x))
data[data['Square']>100]['Bed Room'].value_counts()
```

三居室中塔楼的平均价格为多少

```python
data[(data['Bed Room']=='3') & (data['Building Type']=='塔楼')]["Price"].mean()
```
